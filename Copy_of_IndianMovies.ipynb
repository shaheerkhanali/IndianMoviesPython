```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# meta has imdb_id and movie title and year of release

meta = pd.read_csv("//content/bollywood_meta_2010-2019.csv")
meta.head()
```

```python
# rating has imdb_id and rating

ratings = pd.read_csv("/content/bollywood_ratings_2010-2019.csv")
ratings.head()
```

**1- Check the dataframes basic info.**

```python
meta.info()
```

```python
ratings.info()
```

**2- Check for the duplicate and missing values.**

```python
meta.imdb_id.duplicated()
```

```python
ratings.imdb_rating.duplicated()
```

```python
meta.notna()
```

```python
ratings.notna()
```

**3- Drop duplicates in both dataframes.**

```python
meta=meta.drop_duplicates()
meta
```

```python
ratings=ratings.drop_duplicates()
ratings
```

**4- Check for missing values again and drop them.**

```python
meta=meta.dropna()
meta
```

```python
ratings=ratings.dropna()
ratings
```

**5- Merge the dataframes so that you can have all the info in one dataframe.**

```python
df = meta.merge(ratings, how='inner', on='imdb_id')
df
```

**7- Use Seaborn for visualization and exploratory data analysis.**

```python
sns.catplot(data=df, x='year_of_release', y='imdb_rating', height=8.27, aspect=11.7/8.27)
```

**8- Show the top 10 movies based on rating. Use seaborn for visualizing a barplot for their rating.**

```python
top10=df.nlargest(10, 'imdb_rating')
top10
```

```python
sns.barplot(data=top10, x='imdb_rating', y='title')
```

**9- Show the top 10 movies based on popularity (votes). Use seaborn for visualizing a barplot for their rating.**

```python
topvotes=df.nlargest(10,'imdb_votes')
topvotes
```

```python
sns.barplot(data=topvotes, x='imdb_votes', y='title')
```
